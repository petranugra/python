```python
#2 coding program phyton untuk image augmentation dengan menggunakan objek gambar pribadi
#Augmentasi 2 fungsi untuk menurunkan/meningkatan kecerahan gambar

# Membuka File Image
img = Image.open(image)

plt.subplot(121),plt.imshow(img)
plt.title('Gambar Original'), plt.xticks([]), plt.yticks([])
curr_bri = ImageEnhance.Brightness(img)
new_bri = 0.2 #semakin tinggi nilainya semakin cerah
img_brightened = curr_bri.enhance(new_bri)
plt.subplot(122),plt.imshow(img_brightened) #Augmentasi 2 untuk menurungkan/meningkatan kecerahan gambar
plt.title('Gambar Augmentasi Brightness'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
#1 coding program phyton untuk image segmentation pada penerapan Feature extraction dengan menggunakan objek pribadi

import numpy as np
import cv2


# Muat dua gambar yang akan dicocokkan
img1 = cv2.imread(image, cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread(image, cv2.IMREAD_GRAYSCALE)
img2_flipped = cv2.flip(img2, -1)

# Buat objek ORB
orb = cv2.ORB_create()

# Temukan keypoint dan deskriptor pada kedua gambar
kp1, des1 = orb.detectAndCompute(img1, None)
kp2, des2 = orb.detectAndCompute(img2_flipped, None)

# Buat objek BFMatcher untuk mencocokkan keypoint dengan metode Hamming dan periksa jarak terdekat
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)

# Cocokkan deskriptor
matches = bf.match(des1, des2)

# Urutkan hasil kecocokan berdasarkan jarak (yang paling dekat di awal)
matches = sorted(matches, key=lambda x: x.distance)

# Visualisasikan hasil pencocokan
img_matches = cv2.drawMatches(img1, kp1, img2_flipped, kp2, matches[:10], None, flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS) #matches menentukan banyaknya persamaan

plt.imshow(img_matches)
plt.title('Image Matching 10 Persamaan Gambar Flipped')
plt.show()
```

```python
#1 coding program edge detection dengan menggunakan objek pribadi
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt

# Membuka File Image
img = Image.open(image)

img = cv.imread(image, cv.IMREAD_GRAYSCALE)
edges = cv.Canny(img,100,200) #Mengaplikasikan metode Canny untuk mendeteksi tepi pada gambar. Dua parameter angka (100 dan 200) adalah nilai ambang batas bawah dan atas untuk mendeteksi tepi.

plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Gambar Original (Grayscale)'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Gambar Edge Detection'), plt.xticks([]), plt.yticks([])

plt.show()
```
